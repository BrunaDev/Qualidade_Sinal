```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("dataset.xlsx")
```

```python
df['Date'] = pd.to_datetime(df["Date"])
df = df.sort_values(by="Date")
```

```python
# Cria a coluna de status
df["Status"] = df["Quality"].apply(lambda x: 'OK' if x > 70 else 'NOK')

colunas = ["Site", "Cell", "Date", "Quality", "Status"]

# salva o relatório
df[colunas].to_excel("relatorio_qualidade.xlsx", index=False)

# Filtra
df_cell = df[df["Cell"] == "N1SCTIO053"]
```

```python
# Calcula a média da qualidade por data
df_media = df_cell.groupby("Date")["Quality"].mean()
```

```python
# cria o gráfico
plt.figure(figsize=(10,5))
plt.plot(df_media.index, df_media.values, marker='o', linestyle='-')
plt.xlabel("Data")
plt.ylabel("Média da Qualidade de Sinal")
plt.title("Média Média da Qualidade de Sinal - N1SCTIO053")
plt.grid()
plt.xticks(rotation=45)

plt.savefig("grafico_qualidade.png")
```
